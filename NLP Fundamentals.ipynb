## Feature Engineering
* Two popular methods: 
* (1) Cosine similarity: Between 2 texts is the cosine of the angle between their vector representations.(ex. BoW & TFIDF) 
* (2) Jacard similarity: The ratio of the number of terms common between 2 text documents to the total num of unique terms present in those text.

```python
lemmat=WordNetLemmatizer()
```

```python
pair1=['What you do defines you','Your deeds define you']
pair2=['Once upon a time there lived king','Who is your queen?']
pair3=['He is desperate','Is he not desperate']
```

```python
# Create a function to extract the Jaccard similarity between a pair of sentences.
def Jaccard_sim(text1, text2):
    word1=[lemmat.lemmatize(w.lower()) for w in word_tokenize(text1)]
    word2=[lemmat.lemmatize(w.lower()) for w in word_tokenize(text2)]
    
    nr=len(set(word1).intersection(set(word2)))
    dr=len(set(word1).union(set(word2)))
    jaccard=nr/dr
    return jaccard
```
